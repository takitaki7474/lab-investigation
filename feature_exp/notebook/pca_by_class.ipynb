```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
df = pd.read_pickle("./feature/f_2d_table.pkl")
```

```python
df_by_class = []
for i in range(10):
    df_by_class.append(df[df.label==i])
```

```python
f_list = []
for f in df_by_class[0]["feature"].values:
    f_list.append(f[0])
    
pca = PCA(n_components=2)
X = pca.fit_transform(f_list)
```

```python
plt.scatter(X[:,0],X[:,1])
```
